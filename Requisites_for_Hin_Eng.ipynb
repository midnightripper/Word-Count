```python
import re
import nltk
from nltk.corpus import words

# Get a set of all valid English words
english_words = set(words.words())

# Define a function to count the number of valid English words in a given text
def count_valid_english_words(text):
    # Split the text into individual words
    words = re.findall(r'\b\w+\b', text.lower())
    
    # Count the number of valid English words
    count = sum(1 for word in words if word in english_words)
    
    return count

# Define the text to be searched
text = """
16
00:00:52,560 --> 00:00:55,240
to bring Indian families together.

17
00:01:00,720 --> 00:01:03,000
Inki tagline honi chahiye how are you
"""

# Split the text into individual lines
lines = text.strip().split('\n')

# Loop through the lines and extract the timestamps and text
for i in range(0, len(lines), 3):
    timestamp = lines[i+1]
    text = lines[i+2]
    
    # Count the number of valid English words in the text
    count = count_valid_english_words(text)
    
    # Print the timestamp and word count in seconds format
    start_time, end_time = timestamp.split(' --> ')
    start_seconds = sum(float(x) * 60 ** i for i, x in enumerate(reversed(start_time.split(':'))))
    end_seconds = sum(float(x) * 60 ** i for i, x in enumerate(reversed(end_time.split(':'))))
    print('{:.3f}\t{:.3f}\t{}'.format(start_seconds, end_seconds, count))
```
